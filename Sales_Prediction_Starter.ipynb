Data Wrangling + Feature Engineering

```python
X_train_c['Item_Identifier'].apply(lambda x: x[:2]).value_counts()
```

```python
# Less compute intensive second methods
X_train_c['Item_Identifier'].str
```

```python
X_train_c['Item_Identifier'].str[:2].value_counts()
```

```python
def create_item_type(data_frame):
  data_frame['Item_Type'] = data_frame['Item_Identifier'].str[:2]
  data_frame['Item_Type'] = data_frame['Item_Type'].map({'FD':'Food', 'NC':'Non-Consumable', 'DR':'Drinks'})
  return data_frame
```

```python
X_train_c = create_item_type(X_train_c)
X_train_c.head()
```

```python
X_train_c.isnull().sum()
```

```python
X_train_c[['Item_Identifier', 'Item_Weight']].drop_duplicates().sort_values(by=['Item_Identifier'])
```

```python
ITEM_ID_WEIGHT_PIVOT = X_train_c.pivot_table(index='Item_Identifier', values='Item_Weight').reset_index()
ITEM_ID_WEIGHT_MAPPER = dict(zip(ITEM_ID_WEIGHT_PIVOT['Item_Identifier'], ITEM_ID_WEIGHT_PIVOT['Item_Weight']))
list(ITEM_ID_WEIGHT_MAPPER.items())[:10]
```
